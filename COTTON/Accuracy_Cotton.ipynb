```python
import pandas as pd

# Load the dataset
file_path = 'Accuracy_Cotton.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)

# Fill non-finite values with a default value (e.g., 0)
data['Real-Time'] = pd.to_numeric(data['Real-Time'], errors='coerce').fillna(0).astype(int)
data['Prediction'] = pd.to_numeric(data['Prediction'], errors='coerce').fillna(0).astype(int)

data.info()
```

```python
data.head()
```

```python
Onion_data = data[data['Crop'] == 'Onion']
Corn_data = data[data['Crop'] == 'Corn']
Potato_data = data[data['Crop'] == 'Potato']
Paddy_data = data[data['Crop'] == 'Paddy']
Tomato_data = data[data['Crop'] == 'Tomato']
Cotton_data = data[data['Crop'] == 'Cotton']
```

```python
import matplotlib.pyplot as plt

def plot_data(crop_data, crop_name):
    plt.figure(figsize=(12, 6))
    
    plt.plot(crop_data['Date'], crop_data['Real-Time'], label='Real-Time',linestyle='--', color='blue')
    plt.plot(crop_data['Date'], crop_data['Prediction'], label='Prediction', linestyle='--', color='green')
    
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.title(f'{crop_name} - Real-Time vs Prediction')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    
    # Annotate each point with its value
    for i in range(len(crop_data)):
        plt.text(crop_data['Date'].iloc[i], crop_data['Real-Time'].iloc[i], crop_data['Real-Time'].iloc[i], fontsize=8)
        plt.text(crop_data['Date'].iloc[i], crop_data['Prediction'].iloc[i], crop_data['Prediction'].iloc[i], fontsize=8)
       
    
    plt.show()
```

```python
plot_data(Onion_data, 'Onion')
```

```python
plot_data(Corn_data, 'Corn')
```

```python
plot_data(Potato_data, 'Potato')
```

```python
plot_data(Tomato_data, 'Tomato')
```

```python
plot_data(Paddy_data, 'Paddy')
```

```python
plot_data(Cotton_data, 'Cotton')
```
